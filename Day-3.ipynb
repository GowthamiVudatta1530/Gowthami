```
# prompt: Using dataframe data: i need code for skewness, mean , median,mode,variatio

import pandas as pd
# Assuming your dataframe is named 'data'

# Calculate skewness for numerical columns
skewness_transaction_amount = data['TransactionAmount'].skew()
skewness_customer_age = data['CustomerAge'].skew()
skewness_transaction_duration = data['TransactionDuration'].skew()
skewness_login_attempts = data['LoginAttempts'].skew()
skewness_account_balance = data['AccountBalance'].skew()


print(f"Skewness of TransactionAmount: {skewness_transaction_amount}")
print(f"Skewness of CustomerAge: {skewness_customer_age}")
print(f"Skewness of TransactionDuration: {skewness_transaction_duration}")
print(f"Skewness of LoginAttempts: {skewness_login_attempts}")
print(f"Skewness of AccountBalance: {skewness_account_balance}")

# Calculate mean, median, and mode for numerical columns
def describe_column(column):
    print(f"\nColumn: {column.name}")
    print(f"Mean: {column.mean()}")
    print(f"Median: {column.median()}")
    print(f"Mode: {column.mode()[0]}") # Access the first mode value
    print(f"Variation: {column.var()}")
    # Standard deviation
    print(f"Standard Deviation: {column.std()}")

describe_column(data["TransactionAmount"])
describe_column(data["CustomerAge"])
describe_column(data["TransactionDuration"])
describe_column(data["LoginAttempts"])
describe_column(data["AccountBalance"])
```
