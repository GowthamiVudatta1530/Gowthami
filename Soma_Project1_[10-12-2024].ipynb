# Task-2:-Data cleaning

Task-2.1:-If there are duplicate values remove them

```
synthetic_sales = synthetic_sales.drop_duplicates()
synthetic_sales
```

Task-2.2:-Handle missing values appropriately

```
# prompt: Handle missing values appropriately(drop or impute them)

# Handle missing values by imputation (filling with the mean) for numerical columns
numerical_cols = synthetic_sales.select_dtypes(include=['number']).columns
for col in numerical_cols:
    synthetic_sales[col] = synthetic_sales[col].fillna(synthetic_sales[col].mean())

# Handle missing values by imputation (filling with the most frequent value) for categorical columns
categorical_cols = synthetic_sales.select_dtypes(include=['object']).columns
for col in categorical_cols:
    synthetic_sales[col] = synthetic_sales[col].fillna(synthetic_sales[col].mode()[0])

#Alternatively, you can drop rows with missing values:
#synthetic_sales = synthetic_sales.dropna()

synthetic_sales
```

Task-2.3:-Remaning the columns to more meaning ful names


```
# prompt: Remaning the columns to more meaning ful names

# Rename columns
synthetic_sales = synthetic_sales.rename(columns={
    'Customer ID': 'customer_id',
    'Product Category': 'product_category',
    'Product Name': 'product_name',
    'Sales Channel': 'sales_channel',
    'Order Priority': 'order_priority',
    'Order Date': 'order_date',
    'Order ID': 'order_id',
    'Ship Date': 'ship_date',
    'Units Sold': 'units_sold',
    'Unit Price': 'unit_price',
    'Unit Cost': 'unit_cost',
    'Total Revenue': 'total_revenue',
    'Total Cost': 'total_cost',
    'Total Profit': 'total_profit'
})

synthetic_sales
```





**Duplicate Handling:**

1. **Detection:** The code first identifies duplicate rows using `synthetic_sales.duplicated().sum()`. This calculates the number of rows that are exact copies of earlier rows.

2. **Removal:**  Duplicate rows are removed using `synthetic_sales = synthetic_sales.drop_duplicates()`.  The `drop_duplicates()` method efficiently removes redundant rows, keeping only the first occurrence of each unique row. This is done *after* calculating the number of duplicates, which demonstrates the method's effectiveness.

**Missing Value Handling:**

1. **Detection:** Missing values are identified using `synthetic_sales.isnull().sum()`.  This provides a count of missing values for each column, and the `sum()` of the series shows the total number of missing values in the dataset.

2. **Imputation:** The code handles missing values differently depending on the column's data type:

    * **Numerical Columns:** For numerical columns, missing values are filled with the mean (average) of the respective column using `synthetic_sales[col].fillna(synthetic_sales[col].mean())`.
    * **Categorical Columns:**  Missing values in categorical columns (object type in pandas) are filled with the most frequent value (mode) of the respective column using  `synthetic_sales[col].fillna(synthetic_sales[col].mode()[0])`.

3. **Alternative (commented out):** The code includes a commented-out line `synthetic_sales = synthetic_sales.dropna()`. This shows an alternative strategy, where rows with *any* missing values would be completely removed from the dataset.  This approach might be preferable if missing values are prevalent, however, imputation may be preferred to keep as many data points as possible.


**final summary:**

The code loads a retail sales dataset, cleans it by removing duplicates and handling missing values, and then renames the columns for better readability. Task-3.1:-Plot the distribution of the following discrete features  
1)Quality
2)Category
3)Region

```
import matplotlib.pyplot as plt
import seaborn as sns

def plot_distribution(data, column_name, title):
    """
    Plots the distribution of a discrete feature.

    Args:
        data: The DataFrame containing the feature.
        column_name: The name of the feature column.
        title: The title of the plot.
    """
    plt.figure(figsize=(10, 6))
    if column_name in data.columns:  # Check if column exists
        if data[column_name].dtype == 'object':  # Check if column is categorical (object type)
            sns.countplot(x=column_name, data=data)
        else:  # If column is numerical, create a histogram
            sns.histplot(data[column_name], bins=20)  # Adjust bins as needed
        plt.title(title)
        plt.xlabel(column_name)
        plt.ylabel("Count" if data[column_name].dtype == 'object' else "Frequency")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()
    else:
        print(f"Error: Column '{column_name}' not found in the DataFrame.")

# Plotting distributions for Quantity Sold, Category, and Region
plot_distribution(synthetic_sales, "Quantity Sold", "Distribution of Quantity Sold")  # Assuming 'Quantity Sold' is 'units_sold'
plot_distribution(synthetic_sales, "Category", "Distribution of Category")  # Assuming 'Category' is 'product_category'
plot_distribution(synthetic_sales, "Region", "Distribution of Region")  # Assuming 'Region' is 'sales_channel'
```

# Task-4:-Timestamp Transformation

Task-4.1:-Convert the Order Date column to a datetime format.

```
# prompt: Convert the  Date column to a datetime format.

# Convert 'order_date' column to datetime format
synthetic_sales['Date'] = pd.to_datetime(synthetic_sales['Date'])

# Display the updated DataFrame to confirm the change
synthetic_sales.dtypes
```

Task-4.2:-Create new columns:
1.Day of the week 2.Day of the month
3.Month 4.Year 5.Season (Winter, Spring, Summer, Fall).

```
synthetic_sales['day_of_week'] = synthetic_sales['Date'].dt.dayofweek
synthetic_sales['day_of_month'] = synthetic_sales['Date'].dt.day
synthetic_sales['month'] = synthetic_sales['Date'].dt.month
synthetic_sales['year'] = synthetic_sales['Date'].dt.year

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

synthetic_sales['season'] = synthetic_sales['month'].apply(get_season)

synthetic_sales.head()
```

Task-4.3:-Set the Order Date column as the DataFrame index

```
synthetic_sales = synthetic_sales.set_index('Date')
synthetic_sales
```


1.The code converts the 'Order Date' column (assuming it's named 'Date' in your provided code) to datetime objects.
This is crucial for time-series analysis.  `pd.to_datetime()` handles various date formats.  The updated data type is verified.

2) New columns are derived from the 'Order Date' column:
 1. 'day_of_week':  Represents the day of the week (0 for Monday, 1 for Tuesday, etc.).  Uses `.dt.dayofweek`.
2. 'day_of_month':  Represents the day of the month (1-31).  Uses `.dt.day`.
3. 'month': Represents the month of the year (1-12). Uses `.dt.month`
4. 'year': Represents the year. Uses `.dt.year`
5. 'season':  Categorizes the month into a season (Winter, Spring, Summer, Fall). A helper function `get_season` is defined to map month numbers to seasons.  The `.apply()` method is used to apply the function to each month value.


3)Sets the 'Date' column as the DataFrame's index.  This makes time-based indexing and slicing much more convenient.
`synthetic_sales = synthetic_sales.set_index('Date')`  performs the index change.

# Task-5:-Correlation Analysis

Task-5.1:-Compute the correlation matrix for numerical columns.

```
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical features for correlation analysis
numerical_features = synthetic_sales.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()

# Create a heatmap using seaborn
plt.figure(figsize=(12, 10))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

The heatmap visually represents the correlation matrix using a color scale. Here's a breakdown:

Color Scale: The coolwarm colormap is used by default, where:

Darker red colors indicate strong positive correlations (closer to +1).
Darker blue colors indicate strong negative correlations (closer to -1).
Lighter colors or white indicate weak or no correlations (closer to 0).
Cells: Each cell in the heatmap represents the correlation between two variables. The variables are listed along the x-axis (columns) and y-axis (rows).

Values: The numbers within each cell represent the correlation coefficient between the corresponding variables. These values range from -1 to +1:

+1: Perfect positive correlation (when one variable increases, the other increases proportionally).
-1: Perfect negative correlation (when one variable increases, the other decreases proportionally).
0: No linear correlation (changes in one variable do not predict changes in the other).
Interpreting the Output

Now, let's interpret the output based on the color and values in the heatmap:

Strong Positive Correlation: Look for cells with dark red colors and values close to +1. These indicate a strong positive linear relationship between the two variables. For example, if total_revenue and units_sold have a strong positive correlation, it means that as the number of units sold increases, the total revenue also tends to increase.

Strong Negative Correlation: Look for cells with dark blue colors and values close to -1. This indicates a strong negative linear relationship. For example, if unit_price and units_sold have a strong negative correlation, it suggests that as the unit price increases, the number of units sold tends to decrease.

Weak or No Correlation: Cells with lighter colors or white and values close to 0 indicate a weak or no linear relationship between the variables. This means changes in one variable do not have a predictable effect on the other.
