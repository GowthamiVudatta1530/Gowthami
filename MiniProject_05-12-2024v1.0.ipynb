```
# @title Task-5:- MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC.
# @title Task-5:- MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC.
import pandas as pd

# Assuming your DataFrame is named 'Weather_report'

# Check if 'timestamp' is in columns and not already the index
if 'timestamp' in Weather_report.columns and not Weather_report.index.dtype == 'datetime64[ns]':
    # Convert 'timestamp' to datetime and set as index
    Weather_report['timestamp'] = pd.to_datetime(Weather_report['timestamp'])
    Weather_report = Weather_report.set_index('timestamp')
elif not Weather_report.index.dtype == 'datetime64[ns]':
    # If 'timestamp' is not in columns and index is not datetime, attempt to reset index
    # This assumes 'timestamp' was previously set as index but lost its datetime type
    try:
        Weather_report = Weather_report.reset_index()  # Reset index, 'timestamp' becomes a column
        Weather_report['timestamp'] = pd.to_datetime(Weather_report['timestamp'])  # Convert to datetime
        Weather_report = Weather_report.set_index('timestamp')  # Set 'timestamp' as index again
    except KeyError:
        print("Error: 'timestamp' column not found. Please ensure it exists or the index is a DateTimeIndex.")
        # Handle the error appropriately, e.g., raise an exception or exit

# Now that 'timestamp' is the index and is of datetime type, you can extract features:
Weather_report['day_of_week'] = Weather_report.index.dayofweek  # Monday=0, Sunday=6
Weather_report['day_of_month'] = Weather_report.index.day
Weather_report['hour'] = Weather_report.index.hour
Weather_report['month'] = Weather_report.index.month
Weather_report['year'] = Weather_report.index.year

# For season, define a mapping:
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

Weather_report['season'] = Weather_report['month'].apply(get_season)

# Display the updated DataFrame
# print(Weather_report)
```

```
# @title Task-6:- VISUALIZING THE CORRELATION WITH A HEATMAP

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'Weather_report' and has been processed as in the previous code blocks

# Calculate the correlation matrix, excluding non-numeric columns
# Select only numeric columns for correlation calculation
numeric_columns = Weather_report.select_dtypes(include=np.number).columns
correlation_matrix = Weather_report[numeric_columns].corr()

# Create the heatmap using seaborn
plt.figure(figsize=(12, 10))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

 1.Correlation Matrix: The code calculates the correlation between all pairs of numerical features in your dataset.  A correlation value close to 1 indicates a strong positive relationship (as one variable increases, the other tends to increase), a value close to -1 indicates a strong negative relationship (as one increases, the other tends to decrease), and a value near 0 indicates little to no linear relationship.

 2.Heatmap Visualization:**  This matrix is then visualized as a heatmap.  Each square in the heatmap represents the correlation between two variables. The color intensity of the square shows the strength of the correlation (darker colors mean stronger correlations), and the color (red for positive, blue for negative) shows the direction of the relationship.

 3.Interpretation:By examining the heatmap, you can quickly identify which variables are strongly related.  For example, you might find that temperature and humidity have a strong positive correlation, meaning that when temperature is high, humidity tends to be high as well.  This information can be useful for feature selection, identifying potential redundancies in your data, or understanding underlying patterns in your weather data.

```
# @title Task-7:-VISUALIZING THE CORRELATION OF THE TARGET VARIABLE & THE OTHER FEATURES WITH BARPLO
# Assuming 'cnt' is your target variable
target_variable = 'cnt'

# Calculate correlations with the target variable, but only for numeric columns
numeric_columns = Weather_report.select_dtypes(include=np.number).columns
correlations = Weather_report[numeric_columns].corr()[target_variable].drop(target_variable)  # Exclude self-correlation

# Create a bar plot
plt.figure(figsize=(12, 6))
correlations.plot(kind='bar')
plt.title(f'Correlation of Features with {target_variable}')
plt.xlabel('Features')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```




1.Correlation Calculation:The code calculates the correlation between the target variable ('cnt' in this case) and all other numerical features in your dataset.  It uses the `.corr()` method on the numeric columns of your dataframe and filters the result to include only the correlations with 'cnt'. Self-correlation (the correlation of 'cnt' with itself) is excluded.

2.Bar Plot Creation:A bar plot is created to visualize these correlations. Each bar represents a different feature, and the height of the bar corresponds to the correlation coefficient between that feature and the target variable.

3.Visualization Enhancements:The plot includes a title, axis labels, and rotated x-axis labels to make it more readable. The `plt.tight_layout()` function adjusts the plot layout to prevent labels from overlapping. The plot then is displayed using `plt.show()`.  Positive correlations will be above the 0 line and negative below.  The further from 0 the stronger the relationship.


```
# @title Task-8:-PLOTTING BIKE SHARES OVER TIME BY USING LINEPLOT

# Assuming 'cnt' represents the number of bike shares and your DataFrame is named 'Weather_report'
plt.figure(figsize=(14, 6))
plt.plot(Weather_report.index, Weather_report['cnt'])
plt.xlabel('Time')
plt.ylabel('Number of Bike Shares')
plt.title('Bike Shares Over Time')
plt.grid(True)
plt.show()
```

```
# @title Task-9:-PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)
# prompt: PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Weather_report' DataFrame is already processed as in previous code blocks

# Create a 'year_month' column
Weather_report['year_month'] = Weather_report.index.to_period('M')

# Group by year_month and calculate the mean of 'cnt'
bike_shares_by_month = Weather_report.groupby('year_month')['cnt'].mean()

# Line plot
plt.figure(figsize=(14, 6))
plt.plot(bike_shares_by_month.index.astype(str), bike_shares_by_month.values)  # Convert index to string for plotting
plt.xlabel('Year-Month')
plt.ylabel('Average Bike Shares')
plt.title('Average Bike Shares by Month (Line Plot)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Point plot
plt.figure(figsize=(14, 6))
sns.pointplot(x=bike_shares_by_month.index.astype(str), y=bike_shares_by_month.values)
plt.xlabel('Year-Month')
plt.ylabel('Average Bike Shares')
plt.title('Average Bike Shares by Month (Point Plot)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# Bar plot
plt.figure(figsize=(14, 6))
plt.bar(bike_shares_by_month.index.astype(str), bike_shares_by_month.values)
plt.xlabel('Year-Month')
plt.ylabel('Average Bike Shares')
plt.title('Average Bike Shares by Month (Bar Plot)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
# @title Task-10PLOTTING BIKE SHARES BY HOURS ON (HOLIDAYS, WEEKEND, SEASON):-
# prompt: PLOTTING BIKE SHARES BY HOURS ON (HOLIDAYS, WEEKEND, SEASON)

# Assuming 'Weather_report' DataFrame is already processed as in previous code blocks

# Plotting bike shares by hours on holidays
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='cnt', data=Weather_report[Weather_report['is_holiday'] == 1], label='Holidays')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Bike Shares')
plt.title('Bike Shares by Hour on Holidays')
plt.show()

# Plotting bike shares by hours on weekends
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='cnt', data=Weather_report[Weather_report['is_weekend'] == 1], label='Weekends')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Bike Shares')
plt.title('Bike Shares by Hour on Weekends')
plt.show()

# Plotting bike shares by hours for each season
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='cnt', hue='season', data=Weather_report, ci=None) #ci=None removes confidence intervals
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Bike Shares')
plt.title('Bike Shares by Hour for Each Season')
plt.show()
```
